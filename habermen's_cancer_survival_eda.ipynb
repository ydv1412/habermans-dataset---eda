```python
print("means")
print(np.mean(status_one['axil_nodes']))                #mean auxilary nodes for surviving patients
print(np.mean(status_two['axil_nodes']))                #mean auxilary nodes for dieing patients
print("*************************888")
print("*****************************8")

print("standard deviation")
print(np.std(status_one['axil_nodes']))                #deviation for auxilary nodes for surviving patients
print(np.std(status_two['axil_nodes']))                 #deviation auxilary nodes for dieing patients
```

#### median and MAD and percentiles

```python
###medians
print(np.median(status_one['axil_nodes']))
print(np.median(status_two['axil_nodes']))

print("****************************88")
print("******************************")

print(np.percentile(status_one['axil_nodes'],np.arange(0,100,25)))
print(np.percentile(status_two['axil_nodes'],np.arange(0,100,25)))

print("****************************88")
print("******************************")

print(np.percentile(status_one['axil_nodes'],100))
print(np.percentile(status_two['axil_nodes'],100))
print(np.percentile(status_two['axil_nodes'],40))
print(np.percentile(status_one['axil_nodes'],80))
print("****************************88")
print("******************************")

print(np.percentile(status_one['axil_nodes'],90))
print(np.percentile(status_two['axil_nodes'],90))


print("****************************88")
print("******************************")

from statsmodels import robust
print(robust.mad(status_one['axil_nodes']))
print(robust.mad(status_two['axil_nodes']))
```

conclusion---->
from above percentile we can say that if we set our threshold at 3 we will declare 40 % will survive
and 25% of our survivers as dead
and if we choose 4 as threshold we will declare 50% of our deads as survivers and 20% of our survivers as dead
